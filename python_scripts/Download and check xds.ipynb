```python
import os
import re
import numpy as np
from subprocess import call
import sys
```

```python
call(['wget', 'ftp://ftp.mpimf-heidelberg.mpg.de/pub/kabsch/XDS-OSX_64.tar.gz'])
call(['tar', '-xvf', 'XDS-OSX_64.tar.gz'])

sys.path.append("./XDS-OSX_64")
call('env PATH=./XDS-OSX_64 xds'.split())

call(['rm', 'XDS-OSX_64.tar.gz'])
call(['rm','-r','XDS-OSX_64'])
call(['rm','-r','IDXREF.LP'])
```

```python
def extract_params(params):
    return np.array(map(float,re.split(r'\s+',''.join(params[1:]).strip())))
```

```python
with open('XPARM.XDS_target','r') as f:
    template = f.readlines()
    
with open('XPARM.XDS','r') as f:
    new_version = f.readlines()
    
with open('known formats.txt','r') as f:
    known_formats = f.readlines()
    
if not(new_version[0] in set(known_formats)):
    a=extract_params(template)
    b=extract_params(new_version)

    params_are_almost_equal = np.amax(np.abs(a-b))<1e-5
    if params_are_almost_equal:
        known_formats.append(new_version[0])
        with open('known formats.txt','w') as f:
            f.write('\n'.join(known_formats))
        
        print "Xds updated, XPARM did not change. File with known formats was updated."
        print "New format:"
        print new_version[0]
    else:
        print "Warning! New xds version breaks Meerkat:"
        print new_version[0]
```
